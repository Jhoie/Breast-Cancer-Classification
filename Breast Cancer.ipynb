# MODEL

```python
from sklearn.model_selection import train_test_split
```

```python
# splitting the data into test and train
X = df.drop('diagnosis', axis = 1)  # the features
y = df[['diagnosis']]    # Target outcome
seed = 10 
test_data_size = 0.3 # test data is 30% of entire dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = test_data_size, random_state = seed)
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

# LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()
lg.fit(X_train,y_train)
```

```python
print("Accuracy of Logistic Regression:",lg.score(X_test,y_test))
```

# DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(criterion="entropy")
tree.fit(X_train,y_train)
```

```python
print("Accuracy of Decision Tree:",tree.score(X_test,y_test))
```

# RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators=10, criterion="entropy")
forest.fit(X_train,y_train)
```

```python
print("Accuracy of Random Forest:", forest.score(X_test,y_test))
```

# EVALUATION METRIC
Evaluation metrics are used to measure the quality of the statistical or machine learning model. A confusion matrix gives us a matrix as output and describes the complete performance of the model. I'll use a confusion matrix to measure the performance of the Random Forest Classifier

```python
y_pred = forest.predict(X_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
print(cm)
```

```python

sns.heatmap(cm, annot=True, fmt="g")
```

This analysis observed which features are most helpful in predicting malignant or benign cancer and sees the general trends that aid us in model selection.
